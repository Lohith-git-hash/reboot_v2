```python
# Convert start and end columns to datetime format if they aren't already
df['start'] = pd.to_datetime(df['start'])
df['end'] = pd.to_datetime(df['end'])

# Extract start and end date, and start and end time
df['start_date'] = df['start'].dt.date
df['end_date'] = df['end'].dt.date
df['start_time'] = df['start'].dt.time
df['end_time'] = df['end'].dt.time

# Drop the original start and end columns
df = df.drop(columns=['start', 'end'])

# Display the updated DataFrame
print(df[['start_date', 'end_date', 'start_time', 'end_time']].head())
```

```python
# Remove non-numeric characters and convert to integer
df['Phone Number'] = df['Phone Number'].apply(
    lambda x: pd.NA if pd.isna(x) else int(re.sub(r'\D', '', str(x)))
)

# Convert the column to nullable integer type (Int64)
df['Phone Number'] = df['Phone Number'].astype('Int64')
```

```python
df.info()
```

```python
# Function to safely clean and convert phone numbers
def clean_phone_number(x):
    if pd.isna(x) or x == '':
        return pd.NA  # If NaN or empty, return pd.NA
    else:
        cleaned = re.sub(r'\D', '', str(x))  # Remove non-numeric characters
        return pd.NA if cleaned == '' else int(cleaned)  # Return pd.NA if cleaned is empty

# Apply the cleaning function to the column
df['Business Phone Number (If applicable)'] = df['Business Phone Number (If applicable)'].apply(clean_phone_number)

# Convert the column to nullable integer type (Int64)
df['Business Phone Number (If applicable)'] = df['Business Phone Number (If applicable)'].astype('Int64')
```
